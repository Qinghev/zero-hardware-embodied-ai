# Zero-Hardware Embodied AI: Colab Quick Start

This notebook runs the lite path first. It does not install LeRobot. It downloads the small `lerobot/pusht` parquet/mp4 files and generates episode/action plots.

Project 01 does not need GPU. After the runtime connects, run cells from top to bottom.


```python
from pathlib import Path
import os
import subprocess

repo_url = 'https://github.com/Qinghev/zero-hardware-embodied-ai.git'
workdir = Path('/content/zero-hardware-embodied-ai')

if workdir.exists():
    subprocess.run(['git', '-C', str(workdir), 'pull', '--ff-only'], check=True)
else:
    subprocess.run(['git', 'clone', repo_url, str(workdir)], check=True)

os.chdir(workdir)
print(f'Working directory: {Path.cwd()}')
```
